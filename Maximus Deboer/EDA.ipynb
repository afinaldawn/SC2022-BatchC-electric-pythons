```python
def get_state(x):
    words = x.split(', ')
    state = words[1]
    if state == 'CA':
        return 'CA'
    elif state == 'WA':
        return 'WA'
    elif state == 'NY':
        return 'NY'
    else:
        return 'Other'
```

```python
df['state'] = df['location'].apply(get_state)
df.drop(['location'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
top_15_popular = list(df['company'].value_counts().index)[:15]
def in_popular(x):
    if x in top_15_popular:
        return True
    else:
        return False
    
df['in_top_15'] = df['company'].apply(in_popular)
```

```python
def check_education(master, bach, doc, high, some):
    if master == 1:
        return "Masters"
    elif bach == 1:
        return "Bachelors"
    elif doc == 1:
        return "PHD"
    elif high == 1:
        return 'Highschool'
    elif some == 1:
        return "Some"
    else:
        return "Missing"
```

```python
df['education'] = df.apply(lambda x: check_education(x.Masters_Degree, x.Bachelors_Degree, x.Doctorate_Degree, x.Highschool, x.Some_College), axis=1)
```
